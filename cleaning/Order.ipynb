```python
import unicodecsv

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = read_csv('../data_20180120/order.csv')
```

# เลขที่คำสั่งซื้อ (Order_ID)

```python
import pandas as pd

pd.DataFrame(order)
```

# Change data type

```python
order = read_csv('../data_20180120/order.csv')

df_order = pd.DataFrame(order)

old_order = df_order

import locale
locale.setlocale(locale.LC_ALL, '') 

new_order = old_order[:]
new_order['ORDER_ID'] = old_order['ORDER_ID'].apply(locale.atoi)
new_order['CUSTOMER_ID'] = old_order['CUSTOMER_ID'].apply(locale.atoi)
new_order['BPCODE'] = old_order['BPCODE'].apply(locale.atoi)
new_order['CREATE_DATE'] = pd.to_datetime(old_order['CREATE_DATE'])
new_order['UPDATE_DATE'] = pd.to_datetime(old_order['UPDATE_DATE'])

new_order
```

```python
pd.isnull(new_order).sum() > 0
```

```python
len(new_order)
```

```python
new_order['STATUS'].unique()
```

# Date & time

```python
time_diff = (new_order['UPDATE_DATE']- new_order['CREATE_DATE']).astype('timedelta64[m]')
time_diff
df_time_diff=pd.DataFrame(time_diff, columns=['diff_time'])
join_diff = pd.concat([new_order, df_time_diff], axis=1, join='inner')
join_diff
```

```python
time_join_filt = join_diff[join_diff.diff_time >= 0]
time_join_filt
```

```python
time_join_filt.min()
```

```python
time_join_filt.max()
```

```python
time_cleaned = time_join_filt[(time_join_filt['CREATE_DATE'] > '2015-01-10 16:39:00') & (time_join_filt['CREATE_DATE'] < '2018-01-19')]
time_cleaned
```

```python
time_cleaned.min()
```

```python
time_cleaned.max()
```

```python
del time_cleaned['diff_time']
new_order = time_cleaned
new_order
```

```python
new_order.isnull().sum()
```

```python
new_order.dtypes
```

```python
new_order.to_csv('../data_clean/order.csv', index = False)
```
