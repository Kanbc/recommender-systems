```python
import unicodecsv

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = read_csv('../data_20180120/order.csv')
```

# เลขที่คำสั่งซื้อ (Order_ID)

```python
import pandas as pd

pd.DataFrame(order)
```

# รายการสั่งซื้อที่มีการชำระราคาเสร็จสิ้น (Paid_Success)

```python
order = read_csv('../data_20180120/order.csv')

df_order = pd.DataFrame(order)

order_success = df_order[df_order['STATUS'] == "PAID_SUCCESS"]

import locale
locale.setlocale(locale.LC_ALL, '') 

old_order_success = order_success[:]
old_order_success['ORDER_ID'] = order_success['ORDER_ID'].apply(locale.atoi)
old_order_success['CUSTOMER_ID'] = order_success['CUSTOMER_ID'].apply(locale.atoi)
old_order_success['BPCODE'] = order_success['BPCODE'].apply(locale.atoi)
old_order_success['CREATE_DATE'] = pd.to_datetime(order_success['CREATE_DATE'])
old_order_success['UPDATE_DATE'] = pd.to_datetime(order_success['UPDATE_DATE'])

old_order_success
```

# Check 

```python
time_diff = (old_order_success['UPDATE_DATE']- old_order_success['CREATE_DATE']).astype('timedelta64[m]')
time_diff
df_time_diff=pd.DataFrame(time_diff, columns=['diff_time'])
join_diff = pd.concat([old_order_success, df_time_diff], axis=1, join='inner')
join_diff
```

```python
time_join_filt = join_diff[join_diff.diff_time >= 0]
time_join_filt
```

```python
time_join_filt.min()
```

```python
time_join_filt.max()
```

```python
time_cleaned = time_join_filt[(time_join_filt['CREATE_DATE'] > '2015-01-10 16:39:00') & (time_join_filt['CREATE_DATE'] < '2018-01-01')]
time_cleaned
```

```python
time_cleaned.min()
```

```python
time_cleaned.max()
```

```python
del time_cleaned['diff_time']
new_order_success = time_cleaned
new_order_success
```

```python
new_order_success.dtypes
```

```python
new_order_success['STATUS'].unique()
```

```python
pd.isnull(new_order_success).sum() > 0
```

# Export file order_success

```python
new_order_success.to_csv('../data_clean/order_success.csv')
```
