```python
import unicodecsv

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = read_csv('../data_20180120/order.csv')
```

เลขที่คำสั่งซื้อ (Order_ID)

```python
import pandas as pd

pd.DataFrame(order)
```

รายการสั่งซื้อที่สำเร็จ (Paid_Success)

```python
order = read_csv('../data_20180120/order.csv')

df_order = pd.DataFrame(order)

data_success = df_order[df_order['STATUS'] == "PAID_SUCCESS"]

data_success
```

รายการสั่งซื้อที่คงค้าง (Pending)

```python
order = read_csv('../data_20180120/order.csv')

df_order = pd.DataFrame(order)

data_pending = df_order[df_order['STATUS'] == "PENDING"]

data_pending
```
