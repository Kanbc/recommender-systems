```python
import unicodecsv

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
product = read_csv('../data_20180120/product.csv')
```

```python
import pandas as pd

pd.DataFrame(product)
```
