```python
import unicodecsv
import pandas as pd

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
from datetime import datetime

def bd_to_age(bd_str):
    datetime_object = datetime.strptime(bd_str, '%Y-%m-%d').date()
    now = datetime.now().date()
    delta = (now - datetime_object)
    number_of_days = delta.days
    number_of_years = number_of_days/365
    return number_of_years
```

```python
bd_to_age('1998-01-21')
```

```python
user = read_csv('../data_20180120/customer.csv')
```

```python
users = pd.DataFrame(user)

brith_not_null = users[users['DATE_OF_BIRTH'].notnull() &(users['DATE_OF_BIRTH'] != "")]['DATE_OF_BIRTH']
brith_not_null[0:10]
```

```python
brith_not_null.apply(bd_to_age)
```
