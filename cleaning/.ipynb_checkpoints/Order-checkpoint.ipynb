```python
import unicodecsv

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = read_csv('../data_20180120/order.csv')
```

เลขที่คำสั่งซื้อ (Order_ID)

```python
import pandas as pd

pd.DataFrame(order)
```

รายการสั่งซื้อที่สำเร็จ (Paid_Success)

```python
order = read_csv('../data_20180120/order.csv')

df = pd.DataFrame(order)

df[df['STATUS'] == "PAID_SUCCESS"]
```

รายการสั่งซื้อที่รอ (Pending)

```python
order = read_csv('../data_20180120/order.csv')

df = pd.DataFrame(order)

df[df['STATUS'] == "PENDING"]
```

รายการสั่งซื้อที่ยกเลิก (Cancel)

```python
order = read_csv('../data_20180120/order.csv')

df = pd.DataFrame(order)

df[df['STATUS'] == "CANCEL"]
```
