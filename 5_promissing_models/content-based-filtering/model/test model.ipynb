```python
import unicodecsv
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
product = pd.read_csv('../../2_data/explored/product.csv')
cat = pd.read_csv('../../2_data/explored/cat.csv')
```
