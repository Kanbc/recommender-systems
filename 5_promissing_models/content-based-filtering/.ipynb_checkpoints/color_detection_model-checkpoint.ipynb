```python
import unicodecsv
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

# Detect by image

```python
def get_rgb(image):
    width, height = image.size
    left = (width - (width*0.85))/2
    top = (height - (width*0.65))/2
    right = (width + (width*0.85))/2
    bottom = (height + (width*0.65))/2
    crop = image.crop((left, top, right, bottom))
    try:
        r, g, b = crop.getpixel((55, 55))
    except(IndexError):
        r, g, b = crop.getpixel((35, 35))
    rgb_color = (r, g, b)
    return rgb_color
```

```python
import pandas as pd
import os
import re
import numpy as np
from PIL import Image
import time
start_time = time.time()

indir = '../../2_data/img/'
item_name = []
r_color = []
g_color = []
b_color = []
for root, dirs, filenames in os.walk(indir):
    for f in filenames:
        image = Image.open(indir+f)
        rgb = get_rgb(image)
        r = rgb[0]
        g = rgb[1]
        b = rgb[2]
        r_color.append(r)
        g_color.append(g)
        b_color.append(b)
        file_name = f
        name = re.split(r'\.(?!\d)', file_name)[0]
        item_name.append(name)
        
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import locale

df_item_rgb = pd.DataFrame({'MATNR':item_name, 'R_im':r_color, 'G_im':g_color, 'B_im':b_color})
df_rgb_old = df_item_rgb[['MATNR','R_im','G_im','B_im']]
locale.setlocale(locale.LC_ALL, '')
df_rgb_im = df_rgb_old[:]
df_rgb_im['MATNR'] = df_rgb_old['MATNR'].apply(locale.atoi)
df_rgb_im
```

# Weight data 

```python
df_r = product_rgb_merge[['R_im','R_w']]
df_g = product_rgb_merge[['G_im','G_w']]
df_b = product_rgb_merge[['B_im','B_w']]
```

```python
arr_r = df_r.values.tolist()
arr_g = df_g.values.tolist()
arr_b = df_b.values.tolist()
```

```python
def combine_r(x,y):
    if x != 'None'and y != 'None':
        r_tr = int(round((x * 0.7) + (y * 0.3)))
    elif x == 'None':
        r_tr = y
    elif y == 'None':
        r_tr = x
    else:
        pass
    return r.append(r_tr)

def combine_g(x,y):
    if x != 'None'and y != 'None':
        g_tr = int(round((x * 0.7) + (y * 0.3)))
    elif x == 'None':
        g_tr = y
    elif y == 'None':
        g_tr = x
    else:
        pass
    return g.append(g_tr)

def combine_b(x,y):
    if x != 'None'and y != 'None':
        b_tr = int(round((x * 0.7) + (y * 0.3)))
    elif x == 'None':
        b_tr = y
    elif y == 'None':
        b_tr = x
    else:
        pass
    return b.append(b_tr)
```

```python
r = []
g = []
b = []
for x, y in arr_r:
    combine_r(x,y)
    
for x, y in arr_g:
    combine_g(x,y)

for x, y in arr_b:
    combine_b(x,y)
```

```python
df_r = pd.DataFrame(r, columns=['R'])
df_g = pd.DataFrame(g, columns=['G'])
df_b = pd.DataFrame(b, columns=['B'])
```

```python
df_item_rgb = pd.concat([product_color, df_r, df_g, df_b], axis=1) 
df_item_rgb.drop(df_item_rgb.columns[[1,2,3,4,5]], axis=1, inplace = True)
df_item_rgb
```

```python
df_item_rgb.loc[df_item_rgb.R == 'None', 'R'] = int(round(255*2.1))
df_item_rgb.loc[df_item_rgb.G == 'None', 'G'] = int(round(255*2.1))
df_item_rgb.loc[df_item_rgb.B == 'None', 'B'] = int(round(255*2.1))
```
