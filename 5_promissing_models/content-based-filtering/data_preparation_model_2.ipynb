```python
import unicodecsv
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline

def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
product = pd.read_csv('../../2_data/explored/product.csv')
style = pd.read_csv('../../2_data/explored/style.csv')
cat2 = pd.read_csv('../../2_data/explored/md4t_cat2.csv')
```

```python
product.head()
```

Drop depend on model

```python
product.drop(product.columns[[2,3,4,5,7,8,9,11]], axis=1, inplace = True)
product
```

```python
product_style = pd.merge(product, style,  how='left', left_on=['STY_ID'], right_on = ['STY_ID'])
product_style
```

```python
product_style_cat2 = pd.merge(product_style, cat2,  how='left', left_on=['MVGR4'], right_on = ['NODE'])
product_style_cat2
```

```python
product_style_cat2.rename(columns={'STYLE': 'Style', 'NODE' : 'Node_Cat2', 'LTEXT': 'Cat2'}, inplace=True)
product_style_cat2
```

```python
product_merge = product_style_cat2
product_merge
```

```python
sns.heatmap(product_merge.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_product = product_merge[['MATNR','Style','Cat2']]
df_product
```

```python
df_product_f = df_product.fillna('None')
df_product = df_product_f
sns.heatmap(df_product.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df_cat2 = df_product.drop_duplicates(subset=['Cat2'])
df_cat2.sort_values("Cat2", inplace=True)
df_cat2_idx = df_cat2.reset_index(drop=True)
df_cat2_idx.insert(0, 'Cat2_id', range(1, 1 + len(df_cat2)))
df_cat2_idx.drop(df_cat2_idx.columns[[1,2]], axis=1, inplace = True)
len_cat2_idx = len(df_cat2_idx)
mult = 255 *2.1
len_color = int(round(mult))
constant = len_color - 255
df_cat2_idx.loc[df_cat2_idx.Cat2 == 'None', 'Cat2_id'] = len_cat2_idx + constant
df_cat2_idx
```

```python
df_style = df_product.drop_duplicates(subset=['Style'])
df_style.sort_values("Style", inplace=True)
df_style_idx = df_style.reset_index(drop=True)
df_style_idx.insert(0, 'Style_id', range(1, 1 + len(df_style)))
df_style_idx.drop(df_style_idx.columns[[1,3]], axis=1, inplace = True)
len_style_idx = len(df_style_idx)
df_style_idx.loc[df_style_idx.Style == 'None', 'Style_id'] = len_style_idx + constant
df_style_idx
```

```python
df_product = pd.merge(df_product, df_style_idx,  how='left', left_on=['Style'], right_on = ['Style'])
df_product = pd.merge(df_product, df_cat2_idx,  how='left', left_on=['Cat2'], right_on = ['Cat2'])
df_product_id = df_product
df_product_id
```

```python
df_product_id.to_csv('model/product_id_md2.csv', index=False, encoding='utf-8')
```
