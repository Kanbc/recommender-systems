```python
import unicodecsv
import pandas as pd

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = pd.read_csv('../data_clean/order.csv')
```

```python
import pandas as pd

order = pd.DataFrame(order)

order['CREATE_DATE'] = pd.to_datetime(order['CREATE_DATE'])
order['UPDATE_DATE'] = pd.to_datetime(order['UPDATE_DATE'])

order
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plottprice = tprice
plt.hist(plottprice, normed=True, bins=50)
plt.show()
```
