# Overplot poisson distribution on quantity 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.misc import factorial

data = df_qty['qty']

# the bins should be of integer width, because poisson is an integer distribution
entries, bin_edges, patches = plt.hist(data, bins=11, range=[-0.5, 10.5], normed=True)

bin_middles = 0.5*(bin_edges[1:] + bin_edges[:-1])

def poisson(k, lamb):
    return (lamb**k/factorial(k)) * np.exp(-lamb)

parameters, cov_matrix = curve_fit(poisson, bin_middles, entries) 
x_plot = np.linspace(0,8)

plt.plot(x_plot, poisson(x_plot, *parameters), 'r-', lw=2)
plt.xlabel('$Quantity$')
plt.ylabel('$Probability$')
plt.show()
```

```python
success_order = pd.read_csv('../data_clean/order_success.csv')
num_success_order = len(success_order)
num_success_order


cancel_order = pd.read_csv('../data_clean/order_cancel.csv')
num_cancel_order = len(cancel_order)
#num_cancel_order


pending_order = pd.read_csv('../data_clean/order_pending.csv')
num_pending_order = len(pending_order)
#num_pending_order


total_order = num_success_order + num_cancel_order + num_pending_order
#total_order


percentage_pending = (num_pending_order/total_order) * 100
percentage_cancel = (num_cancel_order/total_order) * 100
percentage_success = (num_success_order/total_order) * 100
```

```python
print(percentage_pending)
print(percentage_cancel)
print(percentage_success)

print ("Success order %.2f%% | Pending order %.2f%% | Cancel order %.2f%%" % (percentage_success, percentage_pending, percentage_cancel))
```

```python
import matplotlib.pyplot as plt
 
# Data to plot
labels = 'Success', 'Cancel', 'Pending'
sizes = [percentage_success, percentage_cancel, percentage_pending]
colors = ['yellowgreen', 'red', 'gold']
explode = (0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.show()
```
