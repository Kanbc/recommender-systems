```python
import unicodecsv
import pandas as pd

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
order = pd.read_csv('../data_clean/order.csv')
```

```python
import pandas as pd

order = pd.DataFrame(order)

order['CREATE_DATE'] = pd.to_datetime(order['CREATE_DATE'])
order['UPDATE_DATE'] = pd.to_datetime(order['UPDATE_DATE'])

order
```

# Plot histogram on total_price

```python
tprice = order.as_matrix(columns=order.columns[9:])

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

std = np.std(tprice) 
mean = np.mean(tprice)


std
```

```python
mean
```

```python
%matplotlib inline
plottprice = tprice
plt.hist(plottprice, normed=True, bins=150)
plt.show()
```
