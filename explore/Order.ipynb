```python
import unicodecsv
import pandas as pd

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

```python
tprice = order.as_matrix(columns=order.columns[9:])

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

std_tprice = np.std(tprice)
mean_tprice = np.mean(tprice)

print(std_tprice)
print(mean_tprice)

tprice
```

```python
cvt_std_tprice = (tprice - mean_tprice) / (len(tprice) - 1)

cvt_std_tprice
```

```python
mean_cvt_tprice = np.mean(cvt_std_tprice)

std_cvt_tprice = np.std(cvt_std_tprice)

print(mean_cvt_tprice)
print(std_cvt_tprice)
```

```python
df_cvt_tprice=pd.DataFrame(cvt_std_tprice, columns=['sd_tprice']) 
df_cvt_tprice
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.misc import factorial

data = df_cvt_tprice['sd_tprice']

# the bins should be of integer width, because poisson is an integer distribution
entries, bin_edges, patches = plt.hist(data, bins=11, range=[-0.5, 10.5], normed=True)

bin_middles = 0.5*(bin_edges[1:] + bin_edges[:-1])

def poisson(k, lamb):
    return (lamb**k/factorial(k)) * np.exp(-lamb)

parameters, cov_matrix = curve_fit(poisson, bin_middles, entries) 
x_plot = np.linspace(0, 7)

plt.plot(x_plot, poisson(x_plot, *parameters), 'r-', lw=2)
plt.show()
```
