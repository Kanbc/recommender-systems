```python
import unicodecsv
import pandas as pd

#read data from CSV file and make it to lists
def read_csv(filename):
    with open(filename, 'rb') as f:
        reader = unicodecsv.DictReader(f)
        return list(reader)
```

# Plot histogram by total_price

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *
from scipy.optimize import curve_fit
from scipy.misc import factorial
from scipy.stats import poisson
from scipy.stats import norm  
%matplotlib inline

data = df_tprice['tprice']

# the bins should be of integer width, because poisson is an integer distribution
entries, bin_edges, patches = plt.hist(data, bins=11, range=[-0.5, 10.5], normed=True)

bin_middles = 0.5*(bin_edges[1:] + bin_edges[:-1])

def poisson(k, lamb):
    return (lamb**k/factorial(k)) * np.exp(-lamb)

parameters, cov_matrix = curve_fit(poisson, bin_middles, entries) 
x_plot = np.linspace(0,20)

plt.plot(x_plot, poisson(x_plot, *parameters), 'r-', lw=2)
plt.show()

```

# Overplot poisson distribution on quantity 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.misc import factorial

data = df_qty['qty']

# the bins should be of integer width, because poisson is an integer distribution
entries, bin_edges, patches = plt.hist(data, bins=11, range=[-0.5, 10.5], normed=True)

bin_middles = 0.5*(bin_edges[1:] + bin_edges[:-1])

def poisson(k, lamb):
    return (lamb**k/factorial(k)) * np.exp(-lamb)

parameters, cov_matrix = curve_fit(poisson, bin_middles, entries) 
x_plot = np.linspace(0,8)

plt.plot(x_plot, poisson(x_plot, *parameters), 'r-', lw=2)
plt.xlabel('$Quantity$')
plt.ylabel('$Probability$')
plt.show()
```
